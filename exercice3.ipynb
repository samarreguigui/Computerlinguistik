```python
#load the data
def load_mails(data_dir):
    data = []
    for label in ["spam", "ham"]:
        folder = data_dir + "/" + label
        for filename in os.listdir(folder):
            with open(folder + "/" + filename, encoding="latin1") as f:
                text = f.read()
                data.append((text, label))
    return data

```

```python
import math
import random
from collections import defaultdict

def model_train(epochs, train_data, classes, learning_rate=0.1):
    # 1. Gewichtsvektor initialisieren
    weight = defaultdict(float)

    # 2. Training Ã¼ber mehrere Epochen
    for n in range(epochs):
        random.shuffle(train_data)  # Daten mischen
        for text, true_class in train_data:

            # a) Merkmalsvektoren berechnen
            # nur einmal pro Klasse
            features_per_class = {}
            for c in classes:
                features_per_class[c] = extract_features(text, c)

            # b) Scores fÃ¼r alle Klassen berechnen
            scores = {}
            for c in classes:
                s = 0.0
                for f, v in features_per_class[c].items():
                    s += weight[(c, f)] * v
                scores[c] = s

            # c) p(Klasse|Mail) berechnen (Softmax)
            max_s = max(scores.values())
            logZ = max_s + math.log(sum(math.exp(scores[c] - max_s) for c in classes))
            probs = {c: math.exp(scores[c] - logZ) for c in classes}

            # d) Gewichte anpassen (Gradientenregel)
            for c in classes:
                for f, v in features_per_class[c].items():
                    # Beobachtete Merkmale (wahre Klasse)
                    if c == true_class:
                        weight[(c, f)] += learning_rate * v
                    # Erwartete Merkmale (alle Klassen)
                    weight[(c, f)] -= learning_rate * probs[c] * v

        print(f"Epoche {n+1} abgeschlossen")

    return weight



```
